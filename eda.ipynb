```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
# read in the data
df = pd.read_csv("data/train.csv")
```

```python
df.head()
```

```python
df.describe()
# the variables are not scaled
```

```python
# check for balance in classes - there are about 10% 1s, so we will have to resample/stratified sample
sum(df['target']) / len(df)
```

```python
# at this point we will do train test split and perform the rest of the eda on the development split
dev_df, holdout_df = train_test_split(df, stratify = df['target'], test_size = 0.2)
```

```python
# holdout_df.to_csv("data/holdout.csv")
# dev_df.to_csv("data/dev.csv")
```

```python
corr_df = dev_df.drop(columns='ID_code').corr() # check correlations
```

```python
corr_df[['target']].iloc[1:][corr_df['target'].iloc[1:] == corr_df['target'].iloc[1:].max()]
 # var 6 with the greatest correlation
                    
```

```python
corr_df[['target']].iloc[1:][corr_df['target'].iloc[1:] == corr_df['target'].iloc[1:].min()]
# var 81 with lowest correlation
```

```python
# split data based on the target column, potentially to look for the variable with greatest difference between the groups
grouped_df = dev_df.groupby('target').apply(lambda x: x.mean())
```

```python
grouped_df.loc["diff"] = np.abs(grouped_df.loc[0] - grouped_df.loc[1])/(grouped_df.loc[0] + grouped_df.loc[1])
```

```python
grouped_df 
```

```python
grouped_df.loc[['diff']].transpose()[grouped_df.loc["diff"] == grouped_df.loc["diff"].max()] # biggest difference
```
